```python
import numpy as np
%matplotlib inline 
import matplotlib.pyplot as plt
import timeit
import numba
#un autre nouveau commentaire
```

```python
d=1000
MAX_ITERATION = 255
XMIN, XMAX, YMIN, YMAX = -1.5, +0.5, -1, +1
```

```python
#initialiser le plan
plan=np.linspace(XMIN,XMAX,d)+1j*(np.transpose(np.linspace(YMAX,YMIN,d)+np.zeros((d,d))))
```

```python
#initialiser la suite des Z
Z = np.zeros((d,d), dtype=complex)
```

```python
#initialiser la fractale
fractale=np.ones((d,d))*255
fractinv=np.zeros((d,d))
```

```python
start_time = timeit.default_timer()
#boucle de calcul
i=1
for i in range(MAX_ITERATION):
   Z=Z**2+plan
   fractale=fractale-np.nan_to_num(np.abs(Z)*np.zeros((d,d)),nan=1)
   fractinv=fractinv+np.nan_to_num(np.abs(Z)*np.zeros((d,d)),nan=1)
elapsed = timeit.default_timer() - start_time
elapsed
```

```python
#affichage
start_time = timeit.default_timer()
array_image=np.nan_to_num(fractale*255/np.max(fractale)).astype("uint8")
fig = plt.figure()
fig.set_size_inches(15,15)
plt.imshow(array_image)
#plt.savefig("/Users/bricemilan/Documents/CQF/fractale3.png", dpi=300)
elapsed = timeit.default_timer() - start_time
elapsed
```

```python
array_image=np.nan_to_num(fractinv*100,nan=255).astype("uint8")
fig = plt.figure()
fig.set_size_inches(15,15)
plt.imshow(array_image)
```

```python
def fractale(d=1000, MAX_ITERATION = 255, XMIN=-1.5, XMAX=+0.5, YMIN=-1, YMAX=+1):
    plan=np.linspace(XMIN,XMAX,d)+1j*(np.transpose(np.linspace(YMAX,YMIN,d)+np.zeros((d,d))))
    Z = np.zeros((d,d), dtype=complex)
    fractale=np.ones((d,d))*255
    i=1
    for i in range(MAX_ITERATION):
        Z=Z**2+plan
        fractale=fractale-np.nan_to_num(np.abs(Z)*np.zeros((d,d)),nan=1)
    array_image=np.nan_to_num(fractale*255/np.max(fractale)).astype("uint8")
    return array_image
```
