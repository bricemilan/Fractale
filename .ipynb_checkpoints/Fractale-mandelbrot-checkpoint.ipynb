```python
import numpy as np
#%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
def fractale(d=1000, MAX_ITERATION = 255, XMIN=-1.5, XMAX=+0.5, YMIN=-1, YMAX=+1):
    plan=np.linspace(XMIN,XMAX,d)+1j*(np.transpose(np.linspace(YMAX,YMIN,d)+np.zeros((d,d))))
    Z = np.zeros((d,d), dtype=complex)
    fractale=np.ones((d,d))*255
    i=1
    for i in range(MAX_ITERATION):
        Z=Z**2+plan
        fractale=fractale-np.nan_to_num(np.abs(Z)*np.zeros((d,d)),nan=1)
    array_image=np.nan_to_num(fractale*255/np.max(fractale)).astype("uint8")
    return array_image
```

```python
plt.imshow(fractale())
```
